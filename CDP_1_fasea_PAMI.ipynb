## Azalpena eta ekuazioa

Modularitatea sare baten propietate bat da eta sarea komunitatetan banatzeko proposamen konkretu bat da. Banaketa egokia izango da komunitateen barruan ertz asko badaude, eta komunitate artean gutxi. Erabiliko dugun modularitate formula goiko paper-eko Clauset, A. et al proposatutakoa da. Izan bedi auzokidetasun matrizeko $A_{vw}$ elementua,  ertzen arteko pisuekin:

$$A_{vw} = \begin{cases}
 & pisua \text{ baldin  $v$ eta $w$ nodoak konektatua daude}  \\ 
 & 0 \text{ bestela } 
\end{cases}$$

Suposa dezagun erpinak komunitatetan banatuta daudela non $v$ $c_v$ komunitatekoa den. Beraz, komunitate barruan dauden nodoen frakzioa honakoa da:

$$\frac{\sum_{vw}A_{vw}\delta(c_v,c_w)}{\sum_{vw}A_{vw}}=\frac{1}{2m}\sum_{vw}A_{vw}\delta(c_v,c_w),\tag{1}$$

non $\delta$ funtzioa $\delta(i, j)$ en emaitza 1 den $i = j$ bada eta 0 bestela, eta $m = \frac{1}{2}\sum_{vw}A_{vw}$ grafoko ertzen pisuen batura den.   

Frakzio hori handia izango da sarearen banaketa on bat daukagunean baina hau ez da nahikoa banaketa ona den jakiteko. Zeren  bere balio handiena hartuko du kasu tribial batean (komunitate bakarra dagoenean eta ertz guztiak 1 direnean). Baina kantitate berdineko ausazko sare bat izanda eta honi espero den balioa kentzen badiogu, modularitatea lortzen dugu:

$$Q=\frac{1}{2m}\sum_{vw}\left[A_{vw}-\frac{k_vk_w}{2m}\right]\delta(c_v,c_w),\tag{2}$$

non $k_v = \sum_{w}A_{vw}$ $gradua$ adierazten duen; hau da, $v$ erpinari lotuta dauden ertzen batura. $k_w$-k berdina esaten digu baina $w$ erpinarekin. Bestalde, konexioak ausaz eginez baina erpinen graduak errespetatuz,  v eta w erpinen artean ertz bat existitzeko probabilitatea $k_{v}k_{w}/2m$ da.

Komunitate barruko ertzen frakzioa ausazko sare batetik espero dezakegunaren berdina bada 0 lortuko dugu bestela 0 baino balio altuago bat lortuko dugu. Beraz, balio altuak lortzen baditugu esan nahi du ondo banatu dugula gure sarea komunitateetan. Gainera praktikan ikusi da 0,3 baino balio altuagoak sarearen komunitate estruktura ona denaren adierazgarri direla. Hortaz, gure helburua balio hau maximizatzea izango da.

Bestalde aipatu behar da modularitatea adierazteko beste era bat dagoela aurrerago ikusiko duguna.

## Pseudokodea

**Sarrera:**
* $partition$ : nodoen partizioa, hiztegi bat izango da non *key* idazlearen izena izango da eta *value* idazle hori dagoen partizioaren zenbakia
* $G$ : grafo bat nodo bakoitza idazle bat izanik eta ertzak idazleen arteko elkarlana adieraziz


**Irteera:** $Q$ : modularitatea

**Hasieraketa:**
> $batu = 0$
>
> $m = \frac{1}{2}\sum_{vw}A_{vw}$
>
> $Q = 0$
>
**Algoritmoa:**
> **for** v in G
>>
>> $k_v = \sum_{w}A_{vw}$
>>
>> **for** w in v  
>>> **if**  $\delta(c_v,c_w) == 1$ **then**
>>>
>>>> batu = batu + $\left[A_{vw}-\frac{k_vk_w}{2m}\right]$
>>
> $Q = \frac{batu}{2m}$

Funtzio orokorrak sortuko ditugu. Zehazki, `sortu_grafoa()` funtzioa sortuko dugu. Funtzio hau emanda ematen ziguten baina moldatu egin dugu funtzioak bukleak sortzen zituelako. Hau da, nodo bat bere buruarekin konektatzen zuen. Honek zentzua ez dauka autore batek ez duelako artikulu bat idazten bere buruarekin baizik eta beste autore batzuekin.

```python
def sortu_grafoa():

    # Datuak irakurri
    # Get data
    connect = sqlite3.connect('./nips-papers/database.sqlite')
    query = """
    SELECT pa.paper_id, pa.author_id, a.name
    FROM paper_authors AS pa JOIN papers AS p ON pa.paper_id = p.id
    JOIN authors as a ON pa.author_id = a.id
    WHERE p.Year BETWEEN '2014' AND '2015'
    """
    df = pd.read_sql(query, connect)

    # Have a look at data
    df.head(10)
    
    # Sortu grafoa
    # Initialize graph
    G = nx.Graph()

    # Transform
    # Autorearen IDa erabili beharrean erabili izena.
    for p, a in df.groupby('paper_id')['name']: 
        for u, v in itertools.combinations(a, 2):
            if u != v:#ez badira nodo berdina
                if G.has_edge(u, v):
                    G[u][v]['weight'] +=1
                else:
                    G.add_edge(u, v, weight=1)
    # Print graph size
    print('Autore kopurua grafoan:', G.number_of_nodes())
    print('Elkarlan kopurua grafoan:', G.number_of_edges())
    
    return G
```

Orain modularitatea kalkulatu dugu goian jarritako pseukodeari jarraituz.

```python
### Helburu-funtzioa
def modularitatea1(G, partizioa, weight='weight'):
    ### Bete hemen 15-20 lerro
    m = G.size(weight=weight)
    batu = 0
    for node in G: #nodo bakoitzeko
        kv = G.degree(node, weight=weight)
        for nbr, eattr in G[node].items(): #nbr = auzokide zehatza; eattr=nbr-ren datuak, adib pisuak
            wt = eattr[weight] #n eta nbr arteko pisuak
            kw = G.degree(nbr, weight=weight)
            if(partizioa[node] == partizioa[nbr]):
                batu += wt - (kv * kw)/(2*m)
    modularitatea = batu/(2*m)
    
    return modularitatea

### Dei orokorrak
G=sortu_grafoa()
## Gure inplementazioa
partition1= dict(zip(G.nodes, soluzioa_1))
partition2= dict(zip(G.nodes, soluzioa_2))

print("Gure inplementazioko modularity:")
print("1. soluzioaren modularitatea1: ", modularitatea1(G, partition1))
print("2. soluzioaren modularitatea1: ", modularitatea1(G, partition2))
print("Community moduluko modularity:")
print("1. soluzioaren modularitatea:", community.modularity(partition1, G))
print("2. soluzioaren modularitatea:", community.modularity(partition2, G))
```

Ikus daitekenez *community* moduluak ematen diguna eta lortzen duguna ez da berdina hau da communty moduluak ez duelako berdin kalkulatzen. Community moduluak honela kalkulatzen du, Clauset, A. et al proposatzen duen moduan hain zuzen:

$$Q=\frac{1}{2m}\sum_{vw}\left[A_{vw}-\frac{k_vk_w}{2m}\right]\sum_i\delta(c_v,i)\delta(c_w,i) \\ 
=\sum_i\left[\frac{1}{2m}\sum_{vw}A_{vw}\delta(c_v,i)\delta(c_w,i) \\
-\frac{1}{2m}\sum_{v}k_{v}\delta(c_v,i)\frac{1}{2m}\sum_{w}k_{w}\delta(c_w,i)\right] \\
=\sum_{i} (e_{ii} - a_i^2).\tag{3}$$

non, 
$$e_{ij}=\frac{1}{2m}\sum_{vw}A_{vw}\delta(c_v,i)\delta(c_w,j),\tag{4}$$

$i$ komunitatetik hasita $j$ komunitatea lotzen duten ertzen pisuen zatikia, eta

$$a_i=\frac{1}{2m}\sum_{v}k_v\delta(c_v,i),\tag{5}$$

i komunitateko nodoetan duten ertzen pisuen zatikia da. Beraz, $\delta(c_v,c_w) = \sum_i\delta(c_v,i)\delta(c_w,i)$, idatzi ezkero,(3) ekuazioa daukagu.  

Hortaz, honela kalkulatuko dugu modularitatea ikusteko ea emaitza berdinak lortzen ditugun:




```python
### Helburu-funtzioa
def modularitatea2(G, partizioa, weight='weight'):
    ### Bete hemen 15-20 lerro
    m = G.size(weight=weight)
    modularitatea = 0
    for i in range(max(partizioa.values())+1):
        a, e = 0,0
        for node in G: #nodo bakoitzeko
            if(partizioa[node] == i):
                kv = G.degree(node, weight=weight)
                a += kv 
                for nbr, eattr in G[node].items(): #nbr = auzokide zehatza; eattr=nbr-ren datuak, adib pisuak
                    wt = eattr[weight] #n eta nbr arteko pisuak
                    if(partizioa[nbr] == i):
                        e += wt 
        modularitatea += e/(2*m)-(a/(2*m))**2
    
    return modularitatea

### Dei orokorrak
G=sortu_grafoa()
## Gure inplementazioa
partition1= dict(zip(G.nodes, soluzioa_1))
partition2= dict(zip(G.nodes, soluzioa_2))

print("Gure inplementazioko modularity:")
print("1. soluzioaren modularitatea: ", modularitatea2(G, partition1))
print("2. soluzioaren modularitatea: ", modularitatea2(G, partition2))
print("Community moduluko modularity:")
print("1. soluzioaren modularitatea:", community.modularity(partition1, G))
print("2. soluzioaren modularitatea:", community.modularity(partition2, G))
```

```python
#Optimoago egiten saitzen:
def kalkulatuHasierakoMAT(G, weight='weight'):
    K = [[0 for i in range(len(G))]for j in range(len(G))]
    A = [[0 for i in range(len(G))]for j in range(len(G))]
    i = 0
    m = G.size(weight=weight)
    degree = {}
    for node in G:
        j = 0
        degree[node] = G.degree(node, weight=weight)
        for nbr, eattr in G[node].items():
            K[i][j] = G.degree(node, weight=weight)*G.degree(nbr, weight=weight)
            A[i][j] = eattr[weight]
            if i == j:
                A[i][j] *= 2
            j += 1
        i += 1
    K = np.array(K)/np.array(2*m)
    A = np.array(A)
    AK = (A-K)/np.array(2*m)
    
    return AK,m,degree

#Gehiago optimizatu beharko litzateke
#Momentuz ez da erabiltzen.
def aldatuPart(part):
    partB = [[0 for i in range(len(part))]for j in range(len(part))]
    i = 0
    for aut1 in part:
        j = 0
        for aut2 in part:
            if part[aut1] == part[aut2]:
                partB[i][j] = 1
            j += 1
        i +=1
    partB = np.array(partB)
    print(partB)
    return partB
    
def modularitatea1OPT(G, partizioa,AK, weight='weight'):
    '''
    partB = aldatuPart(partizioa)
    m = G.size(weight=weight)
    return np.sum(AK*partB)/(2*m)
    '''
    
    batu = 0
    i = 0
    for node in G: #nodo bakoitzeko
        j = 0
        for nbr, _ in G[node].items(): #nbr = auzokide zehatza; eattr=nbr-ren datuak, adib pisuak
            if(partizioa[node] == partizioa[nbr]):
                batu+=AK[i,j]
            j += 1
        i += 1
    return batu

def modularitatea2OPT(G, partizioa, degree, m, weight='weight'):
    a = defaultdict(float)
    e = defaultdict(float)
    modularitatea = 0
    
    for node in G: #nodo bakoitzeko
        c = partizioa[node]
        a[c] += degree[node]
        for nbr, eattr in G[node].items(): #nbr = auzokide zehatza; eattr=nbr-ren datuak, adib pisuak
            if partizioa[nbr] == c:
                e[c]  += eattr[weight]
        
    for i in set(partizioa.values()):
        modularitatea += (e[i] / (2 * m)) - (a[i] / (2 * m)) ** 2
    
    return modularitatea    

```
